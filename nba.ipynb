```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
```

```python
players = pd.DataFrame(players.get_players())
players
```

```python
players[players['full_name']=='Kobe Bryant']
```

```python
df_kobe = playercareerstats.PlayerCareerStats(player_id = '977')
df_kobe = df_kobe.get_data_frames()[0]
df_kobe
```

```python
df_kobe.info()
```

```python
df_kobe['TOT_PTS'] = df_kobe['PTS'].cumsum()
df_kobe['TOT_PTS']
```

```python
df_kobe['PPG'] = df_kobe['PTS'] / df_kobe['GP']
df_kobe['PPG']
```

```python
fig = plt.figure(figsize=(19, 7))

sns.barplot(data=df_kobe[['SEASON_ID', 'PPG']], x='SEASON_ID', y='PPG')

plt.xlabel('Temporada')

fig.show()
```
