```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.static import players
```

```python
players = pd.DataFrame(players.get_players())
players
```

```python
kobe_id = players[players['full_name']=='Kobe Bryant']['id']
```

```python
df_kobe = playercareerstats.PlayerCareerStats(player_id = kobe_id)
df_kobe = df_kobe.get_data_frames()[0]
df_kobe
```

```python
df_kobe.info()
```

```python
df_kobe['PPG'] = df_kobe['PTS'] / df_kobe['GP']
df_kobe['PPG']
```

```python
fig = plt.figure(figsize=(19, 7))

sns.barplot(data=df_kobe[['SEASON_ID', 'PPG']], x='SEASON_ID', y='PPG', palette='viridis')

plt.xlabel('Temporada')

fig.show()
```

```python
df_kobe['TOT_PTS'] = df_kobe['PTS'].cumsum()
df_kobe['TOT_PTS']
```

```python
fig = plt.figure(figsize=(18,5))

sns.barplot(data=df_kobe, x='SEASON_ID', y='TOT_PTS', palette='viridis')

```

```python
from nba_api.stats.static import teams

teams = teams.get_teams()
lakers = [team for team in teams if team['abbreviation'] == 'LAL'][0]
lakers_id = lakers['id']
```

```python
from nba_api.stats.endpoints import teamyearbyyearstats

lakers_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=lakers_id)
lakers_stats = lakers_stats.get_data_frames()[0]
lakers_stats = lakers_stats[(lakers_stats['YEAR'] >= '1996-97') & (lakers_stats['YEAR'] <= '2016-17')]
lakers_stats
```
